```python
"""
The Fuction basicly calculates g^A (mod N) (N need not to be a prime)
Fuction takes modulo N, g and A as a parameters..
returns the result of calculation.
"""
def squareAndMultiply(N, g, A):
    # Setting the values
    a = g
    b = 1
    
    while ( A > 0):
        if( A % 2 == 1):
            b = (b * a) % N
        
        a = a^2 % N
        A = A//2
    return b

#Example use of Function.
print("11^507 mod(1237) =", squareAndMultiply(1237, 11, 507), ", Checking with sage's power_mod(g, A, N):", power_mod(11, 507, 1237))
```

```python
from math import ceil, sqrt
def bsgs(g, h, p):
    '''
    Solve for x in h = g^x mod p given a prime p.
    If p is not prime, you shouldn't use BSGS anyway.
    '''
    N = ceil(sqrt(p - 1))  # phi(p) is p-1 if p is prime

    # Store hashmap of g^{1...m} (mod p). Baby step.
    tbl = {pow(g, i, p): i for i in range(N)}

    # Precompute via Fermat's Little Theorem
    c = pow(g, N * (p - 2), p)

    # Search for an equivalence in the table. Giant step.
    for j in range(N):
        y = (h * pow(c, j, p)) % p
        if y in tbl:
            return j * N + tbl[y]

    # Solution not found
    return None


bsgs(2, 893, 1373)
```

```python
def subsetSumProblem(l, S):
    x_list = []
    x_list_f = []
    is_correct = False
    count = 0
    check = S
    
    for i in reversed(l):
        if S >= i:
            x_list.append(1)
            S = S - i
            count = count + i
        else:
            x_list.append(0)
            
    x_list.reverse()
    
    if check == count:
        is_correct = True
    return x_list, is_correct, count

la = [3, 7, 19, 43, 89, 195]
Sa = 260

lb = [5, 11, 25, 61, 125, 261]
Sb = 408

lc = [2, 5, 12, 28, 60, 131, 257]
Sc = 334

ld = [4, 12, 15, 36, 75, 162]
Sd = 214 

print(subsetSumProblem(la, Sa))
print(subsetSumProblem(lb, Sb))
print(subsetSumProblem(lc, Sc))
print(subsetSumProblem(ld, Sd))
```

```python
#Gausian Lattice Reduction,, Solves 
import numpy as np

def gauss_reduction(v1, v2):
    while True:
        if np.dot(v2, v2) < np.dot(v1, v1):
            v1, v2 = v2, v1 # swap step
            #print(v1, v2)
        m = round( np.dot(v1,v2) / np.dot(v1,v1) )
        if m == 0:
            return (v1, v2)
        v2 = v2 - m*v1 # reduction step

v1 = np.array([66586820, 65354729])
v2 = np.array([6513996, 6393464])

print(gauss_reduction(v1, v2))

va1 = np.array([120670, 110521])
va2 = np.array([323572, 296358])

vb1 = np.array([174748650, 45604569])
vb2 = np.array([35462559, 9254748])

vc1 = np.array([725734520, 613807887])
vc2 = np.array([3433061338, 2903596381])

print(gauss_reduction(va1, va2))
print(gauss_reduction(vb1, vb2))
print(gauss_reduction(vc1, vc2))
```
